```python
import ast

def convert(word):
    lists=[]
    for i in ast.literal_eval(word):
        lists.append(i['name'])
    return lists
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
def transform(obj):
    lists=[]
    counts=0
    for i in ast.literal_eval(obj):
        if counts < 3:
            lists.append(i['name'])
            counts +=1
        else:
            break
    return lists
            
```

```python
movies['cast'] = movies['cast'].apply(transform)
```

```python
movies.head(2)
```

```python
movies['crew'][0]
```

```python
def director(obj):
    lists=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            lists.append(i['name'])
            break
    return lists
```

```python
movies['crew'] = movies['crew'] .apply(director)
```

```python

movies['overview'] = movies['overview'].apply( lambda x : x.split())
```

```python
movies.head()
```

```python
movies['genres'][0]
```

```python
# we have to  join the space in words  (Science Fiction) 
movies['genres'] = movies['genres'].apply( lambda x : [i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply( lambda x : [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply( lambda x : [i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply( lambda x : [i.replace(' ','') for i in x])
```

```python
movies['tags'] = movies['overview']+movies['genres'] + movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movies = movies[['movie_id','title','tags']]
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : ' '.join(x))
new_movies['tags'] = new_movies['tags'].apply(lambda x : x.lower())
```

```python
new_movies.head()
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
stopword=stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(word):
    w=str(word)
    reg= re.sub(r'[^a-zA-Z0-9]', ' ', w)
    split=reg.split()
    lemitize=[lemmatizer.lemmatize(w) for w in split if w not in stopword]
    join=' '.join(lemitize)
    return join


new_movies['tags']  =  new_movies['tags'].apply(preprocess)
```

```python
new_movies['tags'] = new_movies['tags'].apply( lambda x : x.split())
```
